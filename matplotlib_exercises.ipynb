```python
#### combination
x = range(0, 100)

# y = sqrt(x)
y1 = [(n ** .5) for n in x]
# y = x^3
y2 = [(n **3 ) for n in x]
#y = 2^x
y3 = [(2 ** n) for n in x]
# y = 1/(x+1)
y4 = [(1/(n+1)) for n in x]

plt.figure(figsize = (8,8))
plt.suptitle('some different math functions')

plt.subplot(221)
plt.plot(x,y1, color = 'r', ls = '--')

plt.subplot(222)
plt.plot(x,y2, color = 'r', ls = '--')

plt.subplot(223)
plt.plot(x,y3, color = 'r', ls = '--')

plt.subplot(224)
plt.plot(x,y4, color = 'r', ls = '--')

plt.tight_layout()
plt.show()
```
