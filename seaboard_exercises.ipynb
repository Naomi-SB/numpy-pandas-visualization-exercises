## swiss dataset

```python
swiss = data('swiss')
```

```python
data('swiss', show_doc = True)
```

```python
swiss.head()
```

```python
#Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
swiss['is_Catholic'] = swiss['Catholic'] >= 68
```

```python
swiss.head()
```

```python
#Does whether or not a province is Catholic influence fertility?
sns.scatterplot(data = swiss, x = 'Fertility', y=swiss.index, hue = 'is_Catholic')
```

```python
sns.boxplot(x = 'is_Catholic', y = 'Fertility', data = swiss)
```

```python
sns.scatterplot(data = swiss, x = 'Fertility', y = 'Catholic', hue= 'is_Catholic')
```

```python
swiss.drop(columns=['is_Catholic'], inplace = True)
swiss.head(1)
```

```python
#What measure correlates most strongly with fertility?
```

```python
sns.pairplot(data=swiss, hue = 'Fertility', corner = True)
```
